```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
face_classifier = cv2.CascadeClassifier("Haarcascades/haarcascade_frontalface_default.xml")
```

```python
img = cv2.imread("Images/eye_face.jpg")
```

```python
plt.imshow(img)
```

```python
fix_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_image)
```

```python
faces = face_classifier.detectMultiScale(fix_image, 1.3, 5)
```

```python
if faces is ():
    print("No face Detected")
```

```python
def detect_face(fix_image):
    face_rects = face_classifier.detectMultiScale(fix_image)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_image, (x,y), (x+w, y+h), (225,225,225), 40)
        
    return fix_image
```

```python
result = detect_face(fix_image)
plt.imshow(result)
```
